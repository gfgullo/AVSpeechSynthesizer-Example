# Un'esempio di Classificazione

```
import numpy as np
import tensorflow as tf
from tensorflow import keras
import pandas as pd
```

```
csv_url = "https://raw.githubusercontent.com/ProfAI/tf00/master/1%20-%20Le%20Basi%20di%20Tensorflow/students.csv"
df = pd.read_csv(csv_url, sep=)
df.head()
```

```
df_sub = df[["studytime", "failures", "absences"]]
df_sub["promosso"] = df["G3"]>=10
```

```
X = df_sub.drop("promosso", axis=1).values
y = df_sub[["promosso"]].values
```

```
model = keras.models.Sequential([
    keras.layers.Dense(1, input_shape=[3], activation="sigmoid")
])
```

```
model.compile(optimizer="sgd", loss="binary_crossentropy", metrics=["accuracy"])
```

```
model.fit(X, y, epochs=100)
```

```
X_test = [[0, 3, 10], [6, 0, 2], [2, 1, 5]]
model.predict(X_test)
```

```
model.predict_classes(X_test)
```
